## 3. ***Data Wrangling***

In the initial state of our hotel booking analysis, we obtained a dataset with 119,390 rows and 32 columns, representing various aspects of hotel bookings. However, the data had some issues that needed attention. Missing values were present in 'country', 'agent', and 'company' columns, and there were duplicate entries as well. To ensure accurate analysis, we cleaned the data by using the following steps :
1. Remove duplicate rows
2. Handling Missing values
3. Convert columns to appropriate datatypes
4. Adding important columns - total guests and total nights stayed

### Data Wrangling Code

**Step 1 - Removing Duplicate Rows**

````python
# First step is to see the total number of duplicate rows
df[df.duplicated()].shape
````

````python
# Removing Duplicate values
df.drop_duplicates(inplace=True)
````

````python
df.shape
````

**Step 2 - Handling Missing values**

````python
# Total number of missing values in the DataFrame
df.isnull().sum()
````

In our hotel booking analysis, the 'company' and 'agent' columns contain data related to the company and agent numbers associated with each booking. However, it is possible that some customers did not book the hotel through any specific agent or company, resulting in missing or null values in these columns. To handle this scenario, we will replace these null values with 0. This approach allows us to retain the distinction between bookings made through agents or companies and those made without any specific agent or company involvement. By imputing 0 for the missing values, we ensure that the data remains informative and ready for further analysis without compromising the integrity of the original dataset.

Next column with missing value is 'country'. This column represents the country of oriigin of customer. So, We will replace the null_values with the 'Unknown' in the country column

````python
# Replacing the missing_values / null_values with 0 in company & agent columns
df['company'].fillna(0, inplace=True)
df['agent'].fillna(0, inplace=True)

# Replacing the missing_values in country column with 'Unknown'
df['country'].fillna('Unknown', inplace=True)
````

There are 4 null_values in children column....So, We will replace the null values under this column with mean value of children.

````python
df['children'].fillna(df['children'].mean(), inplace = True)
````

````python
#CHecing the all null_values are removed or not
df.isnull().sum()
````

Now, There are some rows with total number of adults, children or babies equal to zero. So we will remove such rows.

````python
# checking the sum of adults, children and babies is zero or not
df[df['adults']+df['babies']+df['children'] == 0].shape
````

There are 166 rows with total number of adults, children or babies equal to zero, So we will remove such rows

````python
df.drop(df[df['adults']+df['babies']+df['children'] == 0].index, inplace = True)
````

**Step 3 - Convert columns to appropriate datatypes**




````python
#Converting datatype of columns 'children', 'company' and 'agent' from float to int.
df['children'] = df['children'].astype(int)
df['company'] = df['company'].astype(int)
df['agent'] = df['agent'].astype(int)
````

````python
# Convert date columns to datetime data type
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
````

**Step 4 - Adding important columns - total guests and total nights stayed**

````python
# Adding total_guests column = adults + children + babies
df['total_guests'] = df['adults'] + df['children'] + df['babies']

# Adding total_nights_stayed = stayed in weekend nights + stayed in week nights
df['total_nights_stayed'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
````

### What all manipulations have you done and insights you found?

In our data wrangling process for the hotel booking analysis, we performed several manipulations to prepare the dataset for in-depth exploration. Here are the key steps we took and the insights we found:

- Remove Duplicate Rows: We identified and removed any duplicate entries in the dataset to ensure data integrity. This step allowed us to focus on unique booking records and avoid any repetition in our analysis.

- Handling Missing Values: We addressed missing values in columns like 'country', 'agent', and 'company'. For 'country', we retained the missing values as they represented cases where customers might not have specified a country during the booking process. However, for 'agent' and 'company', which represented agent and company numbers, respectively, we filled the missing values with 0, as it indicated that those bookings were not made through any specific agent or company.

- Convert Columns to Appropriate Datatypes: We converted the 'children', 'company', and 'agent' columns from float to integer data type, as they represented whole numbers. This conversion allowed us to work with these columns more efficiently in subsequent analyses.

- Adding Important Columns: We introduced two new columns, 'total guests' and 'total nights stayed', to gain deeper insights into booking patterns. By summing up the number of 'adults', 'children', and 'babies', we calculated the total number of guests for each booking. Similarly, by combining 'stays_in_weekend_nights' and 'stays_in_week_nights', we derived the total nights stayed.
